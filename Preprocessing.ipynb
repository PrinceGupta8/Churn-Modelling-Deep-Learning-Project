```python
ohe=OneHotEncoder()
```

```python
geo_encoded=ohe.fit_transform(df[['Geography']])
```

```python
geo_encoded.toarray()
```

```python
geo=pd.DataFrame(geo_encoded.toarray(),columns=ohe.get_feature_names_out(['Geography']))
```

```python
geo
```

```python
df=pd.concat([df,geo],axis=1)
```

```python
import pickle
```

```python
with open('ohe_geo.pkl','wb') as file:
    pickle.dump(ohe,file)
```

```python
df
```

```python
df.drop(columns='Geography',inplace=True)
```

```python
oh=OneHotEncoder()
```

```python
gender_enc=oh.fit_transform(df[['Gender']])
```

```python
with open('ohe_gender.pkl','wb') as file:
    pickle.dump(oh,file)
```

```python
gender_int=pd.DataFrame(gender_enc.toarray(),columns=oh.get_feature_names_out(['Gender']))
```

```python
df=pd.concat([df,gender_int],axis=1)
```

```python
df.drop('Gender',axis=1,inplace=True)
```

```python
df.to_csv('preprocess_churn_modelling.csv')
```
